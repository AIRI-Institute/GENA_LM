We note this trade-off in the revised draft. We are also exploring pruning and distillation to speed up inference. Ultimately, we believe the accuracy gains (especially for non-coding genes) justify the additional computational cost in research settings. We have added these points to the Discussion, to ensure users understand how GENATATOR can be utilized for real genomes and what the current limitations are (e.g., compute requirements, partial accuracy).

## Homology and Training/Test Overlap Concerns

**Reviewer’s comment:** The reviewer is concerned that our training setup (using other species’ genomes) may inadvertently **leak test information** via homologous genes. They note that holding out human chromosomes 8, 20, 21 is a standard split, but adding 39 mammalian species in training means *“many of the gene segmentations in the test dataset might be effectively seen during training as opposed to true scaling behavior.”* In other words, closely related species could contain genes homologous to those in the human test set, confounding the results.

**Response:** We agree that **homology is an important factor** to consider. Our aim was to simulate a realistic scenario where a model is pretrained on as much data as available (including related species) and then applied to a new genome. Indeed, many state-of-the-art gene annotation tools leverage comparative genomics (AUGUSTUS and Tiberius, for example, use parameters/hints derived from multiple species). In this sense, including homologous sequences can be seen as using all available signal for annotation, which is biologically justified – the end goal is accurate gene models, and homology is a valid source of signal.

That said, we acknowledge this **may inflate performance** on held-out human genes if those genes have very closely conserved orthologs in the training set. We have taken some precautions (e.g. using only one chromosome per species for testing unseen species, and human test chromosomes are from a single species), but we did not explicitly remove homologs between training and test. In light of the reviewer’s comment, we will add an analysis of potential homology overlap. Specifically, we will identify any test genes that have close orthologs in the training data and assess if performance on those genes is driving the reported improvements. This analysis will help determine how much of the multi-species training gain is due to true generalization vs. memorizing conserved genes.

Importantly, even if some test genes were seen in another species, the model still has to **generalize across significant genomic divergence**. We note that many of the non-human species in training are quite diverged (e.g., monotremes, marsupials) and any leakage is partial (sequence identity is far from 100%). Moreover, for **practical genome annotation**, using comparative data is standard – if a new species is to be annotated, one would leverage all known genomes. Our results demonstrate that DNA LMs can integrate that information effectively. We will clarify this rationale in the revision, while also tempering claims of “true scaling” and clearly discussing this limitation.

We will incorporate these points into the Limitations section to directly address the shortcomings of our approach. A clearer discussion of these points will strengthen the paper.
