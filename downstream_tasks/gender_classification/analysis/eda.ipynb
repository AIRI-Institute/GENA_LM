```python
for sample in data.keys():
    sample_data = data[sample]
    chr_names = sample_data.keys()
    chr_lengths = {k: sample_data[k].shape[0] for k in chr_names}
    print(sum(chr_lengths.values()))
    autosome_lengths = {k: sample_data[k].shape[0] * 2 for k in chr_names if 'chrY' not in k and 'chrX' not in k}
    autosome_bp = sum(autosome_lengths.values())

    sex_lengths = {k: sample_data[k].shape[0] for k in chr_names if k not in autosome_lengths}
    if len(sex_lengths) == 1:
        sex_lengths = {k: sample_data[k].shape[0] * 2 for k in chr_names if k not in autosome_lengths}
    sex_bp = sum(sex_lengths.values())

    chr_lengths = {**sex_lengths, **autosome_lengths}
    print(sum(chr_lengths.values()))
    total_bp = autosome_bp + sex_bp
    chr_probs = {k: chr_lengths[k] / total_bp for k in chr_lengths}
    
    print(f"Sample {sample}")
    
    for chr, prob in chr_probs.items():
        print(f"Chromosome {chr} probability: {prob}")
            
```
